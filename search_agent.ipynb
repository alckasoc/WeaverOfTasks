# Setup

```python
from langchain_community.chat_models.openai import ChatOpenAI
from langchain_community.tools.openweathermap.tool import OpenWeatherMapQueryRun
from langchain_community.tools.you.tool import YouSearchTool

llm = ChatOpenAI()
```

# Agent

```python
from langchain_experimental.agents.agent_toolkits import create_csv_agent
from langchain.agents.agent_types import AgentType

agent = create_csv_agent(
    llm,
    "data/wildfires_lite.csv",
    verbose=True,
    agent_type=AgentType.OPENAI_FUNCTIONS,
    extra_tools=[YouSearchTool(), OpenWeatherMapQueryRun()]
)
```

```python
agent.run("Which country is the least fire-prone, and what is the weather of that country's largest city for today?")
```
